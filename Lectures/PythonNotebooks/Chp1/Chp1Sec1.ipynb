### <span id="sect-1.1.1"></span>What is a PDE?
---

A ***partial differential equation (PDE)*** is an *equation* involving *partial derivatives*. This is not so informative so let’s break it down a bit.

#### <span id="sect-1.1.1.1"></span>What is a differential equation?

An ***ordinary differential equation (ODE)*** is an equation for a function, which we denote often denote by $u$, which depends on *one* independent variable, which we denote by $t$ (or whatever variable is useful for the problem such as $x$) which involves the independent variable, the function, and derivatives of the function:

$$
F( t, u(t), u'(t), u^{(2)}(t), u^{(3)}(t), \ldots, u^{(m)}(t)) = 0.
$$

This is an example of an ODE of ***degree*** $m$ where $m$ is a highest order of the derivative in the equation. The function $F$ is just a function that relates all the variables together.

For instance, if 

$$
F(x, y, z) = -\sin(x) + y + z,
$$

then $F(t,u(t),u'(t))=0$ defines the ODE

$$
    u'(t) + u(t) = \sin(t), 
$$

which we often rewrite more simply as

$$
    u'+u=\sin(t),
$$

where it is understood that $u$ and its derivatives are functions of $t$ and the derivatives are with respect to the variable $t$.

Solving an equation like this on an interval $t\in [0,T]$ would mean finding a function $t \mapsto u(t) \in \mathbb{R}$ with the property that $u$ and its derivatives intertwine in such a way that this equation is true for all values of $t \in [0,T]$. 

The problem can be enlarged by replacing the real-valued $u$ by a vector-valued one $\mathbf{u}(t)= (u_1 (t), u_2 (t), \dots, u_N (t))$. In this case we usually talk about *system of ODEs*. Even in this situation, the challenge is to
find functions depending upon exactly one variable which, together with their derivatives, satisfy the equation.

#### <span id="sect-1.1.1.2"></span>What is a partial derivative?

When you have function that depends upon several variables, you can differentiate with respect to either variable while holding the other variable constant. This spawns the idea of *partial derivatives*. 

As an example, consider a function depending upon two real variables taking values in the reals:

$$
u: \mathbb{R} ^n \to \mathbb{R}.
$$

For $n=2$, we sometimes visualize a function like this by considering its *graph* viewed as a *surface* in $\mathbb{R}^3$ given by the collection of points

$$
\left\{ (x,y,z) \in {\mathbb{R}^3}: z = u(x,y) \right\}.
$$

We can calculate the derivative with respect to $x$ while holding $y$ fixed. This leads to $u_x$, also expressed as $\partial_x u$,
$\frac{\partial u}{\partial x}$, and $\frac{\partial }{\partial x}$. Similary, we can hold $x$ fixed and differentiate with respect to $y$.
In a space like $\mathbb{R}^3$, we often use the "typical" notation of $x$, $y$, and $z$ to refer to the first, second, and third coordinates.
However, we could also use $x_1$, $x_2$, and $x_3$ to denote the first three coordinates where the subscript denotes which coordinate we mean. 
The advantage of this other notation is that it generalizes to $n$-dimensions and also aligns with how we organize arrays of multi-dimensional data on a computer where a specific column (or row) refers to a specific dimension.

A ***partial differential equation*** is an equation for a function which depends on more than one independent variable which involves the independent variables, the function, and partial derivatives of the function:

$$
F(x,y, u(x,y), u_x (x,y), u_y (x,y), u_{xx} (x,y), u_{xy}
(x,y), u_{yx} (x,y), u_{yy} (x,y)) = 0.
$$

This is an example of a PDE of degree 2. Solving an equation like this
would mean finding a function $(x,y) \to u(x,y)$ with the property that $u$ and is partial derivatives intertwine to satisfy the equation.

As with the ODE case, this problem can be enlarged by replacing the real-valued $u$ by a vector-valued one $\mathbf{u}(\mathbf{x})= (u_1 (\mathbf{x}), u_2 (\mathbf{x}), \dots, u_m (\mathbf{x}))$. In this case we usually talk about a *system of PDEs*. Here, we use $\mathbf{x}\in\mathbb{R}^n$, and the goal is to solve the "$n\times m$" dimensional system of PDEs (i.e., find a mapping from $\mathbb{R}^n$ into $\mathbb{R}^m$ that we call $\mathbf{u}$ that satisfies the system of PDEs).

### <span id="sect-1.1.2"></span>Where do PDEs come from?

PDEs are often referred as *Equations of Mathematical Physics* (or *Mathematical Physics* but it is incorrect as Mathematical Physics is now a separate field of mathematics) because many PDEs arise from different domains of physics (acoustics, optics, elasticity, hydro and aerodynamics, electromagnetism, quantum mechanics, seismology etc).

However PDEs appear in other field of science as well (like quantum chemistry, chemical kinetics); some PDEs are coming from economics and financial mathematics, or computer science.

Many PDEs are originated in other fields of mathematics.

### <span id="sect-1.1.3"></span>Examples of PDEs
(Some are actually systems)

- Simplest First Order Equation $$ u_x = 0; $$

- Transport Equation \begin{equation\*} u_t + c u_x = 0; \end{equation\*}

- $\bar{\partial}$ Equation (complex variables) \begin{equation\*} \bar{\partial}f:=\frac{1}{2}(f\_x+ if\_y) =0\end{equation\*} or as $f= u+iv$ \begin{equation\*} \left\\{\begin{aligned} u_x-v_y=0,\\\\ u_y+v_x=0.\end{aligned}\right.\end{equation\*}

- Laplace’s Equation (in 2D) \begin{equation\*} \Delta u:=u_{xx} + u_{yy} = 0\end{equation\*} or similarly in the higher dimensions;

- Heat Equation \begin{equation\*} u_t = k\Delta u;\end{equation\*}

- Schrödinger Equation (quantum mechanics)\begin{equation\*} i u\_t + \Delta u = 0;\end{equation\*}

- Wave Equation \begin{equation\*} u\_{tt} - c^2\Delta u = 0;\end{equation\*}

- Equation of oscillating rod (with one spatial variable) or plate (with two spatial variables)\begin{equation\*} u\_{tt} + K\Delta^2 u = 0;\end{equation\*}

- Maxwell Equation (electromagnetism)\begin{equation\*} \mathbf{E}\_{t} - c\nabla \times \mathbf{H} = 0,\quad \mathbf{H}\_{t} + c\nabla \times \mathbf{E} = 0,\quad \nabla\cdot\mathbf{E}=\nabla\cdot\mathbf{H}=0;\end{equation\*}

- Dirac Equations (quantum mechanics);

- Elasticity Equation\begin{equation\*}\mathbf{u}_{tt}=\lambda \Delta\mathbf{u}+\mu \nabla(\nabla\cdot\mathbf{u});\end{equation\*}

- Navier-Stokes Equation (hydrodynamics for incompressible liquid)\begin{equation\*} \rho\mathbf{v}\_{t} +  (\mathbf{v}\cdot \nabla ) \rho\mathbf{v} -\nu \Delta \mathbf{v}=-\nabla p,\quad \nabla \cdot \mathbf{v} = 0\end{equation\*} where $\mathbf{v}$ is a velocity and $p$ is the pressure; when viscosity $\nu=0$ we get Euler equation;

- Yang-Mills Equation (elementary particles theory);

- Einstein Equation for General Relativity;
and so on...

**<span id="remark-1.1.1"></span>Remark 1.**

1. Some of these examples are actually not single PDEs are systems of PDEs.

2. In all the above examples, there are *spatial variables* $x,y,z$ and often *time variable* $t$, but it is not necessarily so in all PDEs.

3. Equations could be of different order with respect to different variables and it is important. However if not specified the order of equation is the highest order of the derivatives invoked.

4. In the Textbook we will deal mainly with the wave equation (Chapter , heat equation and Laplace equation in their simplest forms.

#### Emphasizing some notation and terminology

- We typically use $u$ to denote the unknown function in a differential equation.

- Typically, we refer to $u$ as the *state variable* since it is used to represent a particular quantifiable aspect (i.e. a state) of a physical system that is modeled by the PDE or ODE. 

  For example, $u$ may represent a contaminant concentration at some particular point in space-time in a transport equation used to model how a contaminant spreads in the subsurface. 

- The variables $x, y,$ and $z$ are typically reserved for representing the usual three-dimensional spatial variables (assuming that the typical Cartesian coordinate system is appropriate for the model and not spherical or cylindrical coordinates).

- We will denote the spatial domain $\Omega$ to represent all the spatial points for which a differential equation is studied, and $\partial\Omega$ will denote the boundary of this domain.

- The variable $t$ is typically reserved for the temporal variable.

- The space-time domain for which the differential equation is to be solved is simply referred to as the *domain*.

- Any coefficients in a differential equation are typically referred to as *parameters*.

- It is common to write $L$ to denote a differential operator so that the differential equation can be written compactly as
<br>
$$
L(u) = f,
$$
<br>
where (the forcing function) $f$ only depends upon the independent domain variables and not the state variable. 
<br><br>
Usually we specify the domain explicitly along with the differential equation. 

- The function $f$ along with any boundary or initial conditions that are specified are usually referred to as the *data*.

  We sometimes refer to $f$ and/or boundary data as source terms. 

- If $f\equiv 0$, then the differential equation is called ***homogeneous***.

- If $f\neq 0$, then the differential equation is called ***nonhomogeneous***.

- Linear differential operators are of particular interest and are used to describe some of the most studied differential equations. 

> A differential operator $L$ is ***linear*** if for all functions $u$ and $v$ (for which it makes sense to evaluate $L(u)$ and $L(v)$) and for all constants $\alpha$ and $\beta$, we have
<br><br>
$$
  L(\alpha u + \beta v) = \alpha L(u) + \beta L(v).
$$
<br>
Some examples of $L$ include $L=\partial_t - \partial_{xx}$ (which defines the heat equation on a 1-dimensional spatial domain modeling the temperature distribution on a thin rod) and $L=\partial_{tt} - \partial_{xx}$ (which defines the wave equation on a 1-dimensional spatial domain modeling oscillations of a string in the absence of friction). 

- A differential operator is ***nonlinear*** if it is not linear.
